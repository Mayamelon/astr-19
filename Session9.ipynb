This program creates a histrogram graph of 1000 random numbers

This cell imports matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

This cell creates an array of 1000 random numbers, each being between 0 and 1. It then specifies 100 bins and plots a histogram. It then saves the figure as a pdf

```python
x = np.random.default_rng().random(1000)
bins = np.linspace(0,1,100)
plt.hist(x, bins=bins, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("Session9.pdf", bbox_inches="tight", facecolor="white")
```
