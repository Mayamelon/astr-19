This program creates a histogram graph of a logarithmic normal distribution

This cell imports matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

This cell creates the random distribution using 1000 elements. It defines 100 bins from 0 to 10. It then plots this graph, labels the axes, and saves the figure as a pdf

```python
x = np.random.lognormal(size=1000)
bins = np.linspace(0,10,100)
plt.hist(x, bins=bins, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("Session10.pdf", bbox_inches="tight", facecolor="white")
```
